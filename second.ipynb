## Предобработка данных

Определение и заполнение пропущенных значений

Заменяем тип данных на целый в столбце last_price

```python
df['last_price']=df['last_price'].astype('int64')
```

Определим пропущенные значения, отсортируем по убыванию и выведем в процентах

```python
pd.DataFrame(df.isna().mean().round(4)*100).sort_values(by=0)[::-1].style.background_gradient('coolwarm')
```

В floors_total (всего этажей в доме) 86 пустых записей, в locality_name (название населённого пункта) 49 значений, учитывая соотношение к общему количеству и то, что название у населенного пункта еще может не быть, удалим эти строки.

```python
df=df.dropna(subset=['locality_name','floors_total'])
```



В balcony (число балконов) 11475 пропущенных значений, учитывая, что сейчас многие дома строяться без них, принимаем, что балконов нет. Заменяем на 0. И меняем тип данных на целый.

```python
df['balcony']=df['balcony'].fillna(value=0)
df['balcony']=df['balcony'].astype('int64')
```

Значения days_exposition (сколько дней было размещено объявление (от публикации до снятия) будем считать как медианную времени экспозиции и общей площади.

```python
days_exposition_group = df.groupby('days_exposition')['total_area'].median()  
df['days_exposition']=df['days_exposition'].fillna(days_exposition_group, inplace=False)
```

Проверяем, что получилось, видим, что стало немного меньше (3050 против 3181 до этого). Остальное заменяем нулями, взять значение нам больше не откуда.

```python
df['days_exposition'] = df['days_exposition'].fillna(0).astype('int')
```

В этой части задания высоту потолков заменим на среднее значение, так мы увидели, что 75% - 2,8 м, а среднее 2,77. Выбросы (минимальное значение 1 метр, что не может быть, и 100 метров будем отсекать на следующем этапе при построении графиков).

```python
ceiling_height_med=df['ceiling_height'].median()
df['ceiling_height'].fillna(ceiling_height_med,inplace=True)
```

В типе квартир "студия" нет выделенной кухни, меняем это значение на 0

```python
df.loc[df['studio'] == 1, 'kitchen_area'] = 0 
```

```python
kitchen_area_med=df.groupby('total_area')['kitchen_area'].median()
for total_area in kitchen_area_med.index:
    df.loc[(df['kitchen_area'].isna()) & (df['total_area'] == total_area), 'kitchen_area'] = kitchen_area_med.loc[total_area]
df['kitchen_area']=df['kitchen_area'].fillna(df['kitchen_area'].mean()) # оставшиеся заменяем по медиане
```

Расстояние до airports_nearest (аэропорта) заполним по медианному значению. Так как эти данные мы получали автоматически ошибка техническая.

```python
df['airports_nearest']=df['airports_nearest'].fillna(df['airports_nearest'].mean())
```

Расстояние до cityCenters_nearest (расстояние до центра города (м)), как и в случае с расстоянием до аэропорта будем смотреть по медианной, ошибка также техническая.

```python
df['cityCenters_nearest']=df['cityCenters_nearest'].fillna(df['cityCenters_nearest'].mean())
```

Разберемся со столбцом апартаменты:

```python
df['is_apartment'].value_counts() # смотрим какие значения есть в столбце, тип целочисленный
```

```python
df['is_apartment']=df['is_apartment'].fillna(False) # изменяем тип на логический
df['is_apartment'].value_counts()  # снова смотрим. Сделали такую замену потому, что иначе класс апартаменты (20855 к 2710) во много раз первосходил бы количесвто квартир и социальная инфраструктура не выдержала бы.
```

Число parks_around3000 (число парков в радиусе 3 км), ponds_around3000 (число водоёмов в радиусе 3 км), parks_nearest (расстояние до ближайшего парка (м)) и ponds_nearest (расстояние до ближайшего водоёма (м)) возьмеме как среднее

```python
df['parks_around3000']=df['parks_around3000'].fillna(df['parks_around3000'].mean())
df['ponds_around3000']=df['ponds_around3000'].fillna(df['ponds_around3000'].mean())
df['parks_nearest']=df['parks_nearest'].fillna(df['parks_nearest'].mean()) # так как какой-то парк должен быть в каком-то расстоянии 
df['ponds_nearest']=df['ponds_nearest'].fillna(df['ponds_nearest'].mean()) # так как какой-то водоем должен быть в каком-то расстоянии
```

Найдем жилую площадь как медиану к количеству комнат

```python
living_area_group = df.groupby('rooms')['living_area'].median()
for rooms in living_area_group.index:
    df.loc[(df['living_area'].isna()) & (df['rooms'] == rooms), 'living_area'] = living_area_group.loc[rooms]
```

```python
df.isnull().sum().sort_values(ascending=False)
```

## Расчёты и добавление результатов в таблицу

Посчитаем и добавим в таблицу цену квадратного метра. Цену квадратного метра будем искать как отношение площади квартиры к стоимости. Новый столбец назовем 'price_m2'. Сделаем значение целым.

```python
df['price_m2'] = df['last_price']/df['total_area']
df['price_m2'] = df['price_m2'].astype('int64')
```

Добавим в таблицу столбец с датой публикации объявления. Для этого преобразуем first_day_exposition и перезапишем его в новом виде

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Посчитаем и добавим в таблицу столбец floor_which, содержащий этаж квартиры по виду — первый, последний, другой.

```python
def floor_which(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'Первый'
    elif floor == floors_total:
        return 'Последний'
    else:                            
        return 'Другой'
   
df['floor_which'] = df.apply(floor_which, axis = 1)
```

Посчитаем и добавим в таблицу соотношение жилой и общей площади, а также отношение площади кухни к общей

```python
df['living_area/total_area'] = df['living_area']/df['total_area']     # соотношение жилой площади и общей площади
df['kitchen_area/total_area'] = df['kitchen_area']/df['total_area']   # соотношение площади кухни и общей площади
```

Задания по 3-ей части выполнены, переходим к следующей части.
